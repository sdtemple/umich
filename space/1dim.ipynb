```python
import numpy as np
from scipy.special import kv, gamma
from variogram import *
import matplotlib.pyplot as plt
```

```python
group1 = np.random.uniform(-10,15,100)
group2 = np.random.uniform(10,35,100)
group3 = np.random.uniform(30,55,100)
groups = np.concatenate((group1,group2,group3))
distances = (groups[np.newaxis,:] - groups[:,np.newaxis])
absolute_distances = np.abs(distances)
```

```python
plt.hist(groups,30,edgecolor='k')
plt.show()
```

```python
sph = spherical_family(absolute_distances,5)
plt.imshow(sph, cmap='inferno_r', interpolation='nearest')
plt.colorbar()
plt.title('Spherical family')
plt.show()
```

```python
sph = powered_exponential_family(absolute_distances,5,1)
plt.imshow(sph, cmap='inferno_r', interpolation='nearest')
plt.colorbar()
plt.title("Powered exponential family")
plt.show()
```

```python
sph = matern_family(absolute_distances,5,1.5)
plt.imshow(sph, cmap='inferno_r', interpolation='nearest')
plt.colorbar()
plt.title("Matern family")
plt.show()
```
